```python
# identify percentage of missing value
missing_percentage = df.isnull().sum() * 100 / len(df)
missing_percentage

# ringkasan keputusan untuk missing value
def missing_decision(pct):
    if pct < 5:
        return "Safe to Remove"
    elif 5 <= pct < 20:
        return "Consider the Context"
    elif 20 <= pct <= 40:
        return "Better to Impute"
    else:
        return "Consider dropping the column"

missing_summary = pd.DataFrame({'Persentase Missing (%)': missing_percentage})
missing_summary['Keputusan'] = missing_summary['Persentase Missing (%)'].apply(missing_decision)
missing_summary = missing_summary.sort_values(by='Persentase Missing (%)',ascending=False)
missing_summary
```

```python
# identify outliers

def detect_outliers_iqr(df):
    outlier_summary = {}
    total_rows = df.shape[0]
    
    numeric_cols = df.select_dtypes(include='number').columns
    
    for col in numeric_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1

        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count = outliers.shape[0]
        outlier_percentage = (outlier_count / total_rows) * 100

        outlier_summary[col] = {
            'lower_bound': lower_bound,
            'upper_bound': upper_bound,
            'outlier_count': outlier_count,
            'outlier_percentage (%)': round(outlier_percentage, 2)
        }

    return pd.DataFrame(outlier_summary).T


outlier_summary = detect_outliers_iqr(df)
outlier_summary
```

```python
# Select continuous numeric variables (float64) for skewness analysis
continuous_cols = df.select_dtypes(include='float64').columns

df[continuous_cols].skew().sort_values(ascending=False)
```

```python
# Histogram for Continuous Variables (float64 only)

import matplotlib.pyplot as plt
import seaborn as sns

float_cols = df.select_dtypes(include='float64').columns

plt.figure(figsize=(18, 10))

for i, col in enumerate(float_cols, 1):
    plt.subplot(4, 3, i)
    sns.histplot(
        df[col],
        kde=True,
        color='teal',
        edgecolor='black',
        alpha=0.7
    )
    plt.title(col, fontsize=11)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Boxplot for Continuous Variables

plt.figure(figsize=(18, 10))

for i, col in enumerate(float_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(
        x=df[col],
        color='teal',
        flierprops=dict(
        marker='o',
        markerfacecolor='red',
        markersize=5,
        linestyle='none'
    )
    )
    plt.title(col, fontsize=11, fontweight='bold')
    plt.xlabel("")

plt.tight_layout()
plt.show()
```

```python
# Countplot for Categorical Integer Variables

int_cols = df.select_dtypes(include='int64').columns

# Filter integer yang kemungkinan kategorikal (unique kecil)
cat_int_cols = [col for col in int_cols if df[col].nunique() < 20]

plt.figure(figsize=(18, 8))

for i, col in enumerate(cat_int_cols, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=df[col], hue=df[col], palette='viridis', legend=False)
    plt.title(col, fontsize=11)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

# Handling Issue

```python
# Handling missing values

# Create state flags
df_cleaning['is_purchased'] = df_cleaning['order_purchase_timestamp'].notna()
df_cleaning['is_approved'] = df_cleaning['order_approved_at'].notna()
df_cleaning['is_delivered'] = df_cleaning['order_delivered_timestamp'].notna()

# Fill missing product category (safe to fill)
df_cleaning['product_category_name'] = df_cleaning['product_category_name'].fillna('Unknown')

# Fill numeric size/weight with median
float_cols = [
    'product_weight_g',
    'product_width_cm',
    'product_height_cm',
    'product_length_cm'
]

for col in float_cols:
    df_cleaning[col] = df_cleaning[col].fillna(df_cleaning[col].median())
```

```python
df_cleaning.isnull().sum()
```

```python
# handling duplicate

df_cleaning = df_cleaning.drop_duplicates(keep='first')
```

```python
df_cleaning.duplicated().sum()
```

# Standardize Data Types

```python
# 1. Define datetime columns
datetime_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_timestamp',
    'order_estimated_delivery_date'
]

# 2. Convert datetime columns
df_cleaning[datetime_cols] = df_cleaning[datetime_cols].apply(pd.to_datetime)

# 3. Convert remaining object columns to category
object_cols = df_cleaning.select_dtypes(include='object').columns

df_cleaning[object_cols] = df_cleaning[object_cols].astype('category')
```
