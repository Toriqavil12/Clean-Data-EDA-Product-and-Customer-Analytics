```python
# identify percentage of missing value
missing_percentage = df.isnull().sum() * 100 / len(df)
missing_percentage

# ringkasan keputusan untuk missing value
def missing_decision(pct):
    if pct < 5:
        return "Safe to Remove"
    elif 5 <= pct < 20:
        return "Consider the Context"
    elif 20 <= pct <= 40:
        return "Better to Impute"
    else:
        return "Consider dropping the column"

missing_summary = pd.DataFrame({'Persentase Missing (%)': missing_percentage})
missing_summary['Keputusan'] = missing_summary['Persentase Missing (%)'].apply(missing_decision)
missing_summary = missing_summary.sort_values(by='Persentase Missing (%)',ascending=False)
missing_summary
```

```python
# identify unique value

unique_summary = pd.DataFrame({
    'n_unique': df.nunique(),
    'unique_values': df.apply(lambda x: x.unique())
})

unique_summary
```

```python
# identify outliers

def detect_outliers_iqr(df):
    outlier_summary = {}
    total_rows = df.shape[0]
    
    numeric_cols = df.select_dtypes(include='number').columns
    
    for col in numeric_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1

        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count = outliers.shape[0]
        outlier_percentage = (outlier_count / total_rows) * 100

        outlier_summary[col] = {
            'lower_bound': lower_bound,
            'upper_bound': upper_bound,
            'outlier_count': outlier_count,
            'outlier_percentage (%)': round(outlier_percentage, 2)
        }

    return pd.DataFrame(outlier_summary).T


outlier_summary = detect_outliers_iqr(df)
outlier_summary
```

```python
# Histogram for Continuous Variables (float64 only)

import matplotlib.pyplot as plt
import seaborn as sns

float_cols = df.select_dtypes(include='float64').columns

plt.figure(figsize=(6,4))

for i, col in enumerate(float_cols, 1):
    plt.subplot(1, 1, i)
    sns.histplot(
        df[col],
        kde=True,
        color='teal',
        edgecolor='black',
        alpha=0.7
    )
    plt.title(col, fontsize=11)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

# Handling Issue

```python
# handling missing value

# Fill missing values of arrival delay in minutes with median
float_cols = ['Arrival Delay in Minutes']

for col in float_cols:
    df_cleaning_1[col] = df_cleaning_1[col].fillna(df_cleaning_1[col].median())
```

```python
# handling inconcistency data

df_cleaning_1['Customer Type'] = (
    df_cleaning_1['Customer Type']
    .str.strip()
    .str.title()
)

df_cleaning_1['Type of Travel'] = (
    df_cleaning_1['Type of Travel']
    .str.strip()
    .str.title()
)
```

# Standardize Data Types

```python
# Convert all object columns to category
object_cols = df_cleaning_1.select_dtypes(include='object').columns

df_cleaning_1[object_cols] = df_cleaning_1[object_cols].astype('category')
```
